```python
# Importamos librerías
import gdown
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
#Importamos dataset

# Reemplaza con el ID del archivo en Google Drive
file_id = '1k9tBLaVd-7B1w_kxWsZu0bpvpWd3xp-q'
dwn_url = f'https://drive.google.com/uc?export=download&id={file_id}'

# Descargar el archivo con gdown
output = 'housing.csv'
gdown.download(dwn_url, output, quiet=False)

# Leer el CSV usando pandas
data = pd.read_csv(output)

# Mostrar los primeros registros del dataset
data.head()
```

```python
# Dataset de entrenamiento y prueba
x = data.drop(['median_house_value', 'ocean_proximity'], axis=1)
y = data['median_house_value']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#entrenar modelo

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#Eliminar valores nulos
x_test = x_test.dropna()
y_test = y_test[x_test.index]

#Comparar predicción con resultado real
y_pred = model.predict(x_test)

data_pred = pd.DataFrame({'Actual Values': y_test, 'Predicted Values:': y_pred})
data_pred
```

a)	¿Qué características influyen más en el valor de una casa?

```python
correlation_matrix = data.drop(columns=['ocean_proximity']).corr()
print(correlation_matrix['median_house_value'].sort_values(ascending=False))
```

```python
# Crear un DataFrame con los coeficientes del modelo
coefficients = pd.DataFrame(model.coef_, x.columns, columns=['Coeficiente'])

# Ordenar los coeficientes de mayor a menor
print(coefficients.sort_values(by='Coeficiente', ascending=False))
```

b)	¿Cuál es la precisión del algoritmo generado?

```python
r2 = r2_score(y_test, y_pred)
print(f"Puntaje R²: {r2}")
```

c)	Usar las métricas de evaluación del algoritmo: error absoluto medio, error cuadrático medio, puntaje r al cuadrado, la raíz del error cuadrático medio, para evaluar el modelo y explicar su interpretación.

```python
# Calcular el Error Absoluto Medio (MAE)
mae = mean_absolute_error(y_test, y_pred)
print(f"Error Absoluto Medio (MAE): {mae}")

# Calcular el Error Cuadrático Medio (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Error Cuadrático Medio (MSE): {mse}")

# Calcular la Raíz del Error Cuadrático Medio (RMSE)
rmse = np.sqrt(mse)
print(f"Raíz del Error Cuadrático Medio (RMSE): {rmse}")

# Calcular el Puntaje R²
r2 = r2_score(y_test, y_pred)
print(f"Puntaje R²: {r2}")
```
