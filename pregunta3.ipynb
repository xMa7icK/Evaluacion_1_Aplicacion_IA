```python
file_id = '1id64cZMHXBJmop0UN0z1RYtUKd-0KX7O'
dwn_url = f'https://drive.google.com/uc?export=download&id={file_id}'

# Descargar el archivo con gdown
output = 'rating.csv'
gdown.download(dwn_url, output, quiet=False)

# Leer el CSV usando pandas
df_calificaciones = pd.read_csv(output)

# Mostrar los primeros registros del dataset
df_calificaciones.head()
```

a)	¿Cuál película recomendarían si se quiere ver una película de terror? 

```python
# Filtrar las películas de terror
df_terror = movie[movie['genres'].str.contains('Horror')]

# Obtener las calificaciones de las películas de terror
calificaciones_terror = df_calificaciones[df_calificaciones['movieId'].isin(df_terror['movieId'])]

# Agrupar por 'movieId' y contar cuántas veces ha sido calificada cada película de terror
conteo_calificaciones_terror = calificaciones_terror.groupby('movieId').size()

# Ordenar las películas de terror por número de calificaciones (de más a menos)
peliculas_terror_populares = conteo_calificaciones_terror.sort_values(ascending=False).index

# Obtener los detalles de las películas más populares de terror
detalles_peliculas_terror_populares = movie[movie['movieId'].isin(peliculas_terror_populares)]

# Mostrar las primeras películas de terror más recomendadas
detalles_peliculas_terror_populares.head()
```

b)	¿Qué película recomendarían si mi última película fue Toy Story? 

```python
# Encontrar el ID de Toy Story
id_toy_story = movie[movie['title'] == 'Toy Story (1995)']['movieId'].values[0]
id_toy_story
```

```python
# Filtrar usuarios que vieron Toy Story
usuarios_que_calificaron_toy_story = df_calificaciones[df_calificaciones['movieId'] == id_toy_story]['userId'].unique()
usuarios_que_calificaron_toy_story
```

```python
# Encontrar otras películas que estos usuarios hayan visto
otras_calificaciones = df_calificaciones[df_calificaciones['userId'].isin(usuarios_que_calificaron_toy_story)]
otras_calificaciones
```

```python
# Agrupar por película y contar cuántas veces ha sido calificada
conteo_peliculas = otras_calificaciones.groupby('movieId').size()
conteo_peliculas
```

```python
# Ordenar por número de calificaciones
peliculas_recomendadas = conteo_peliculas.sort_values(ascending=False).index
peliculas_recomendadas
```

```python
# Verificación del ID de Toy Story
print(f"ID de Toy Story: {id_toy_story}")

# Filtrar Toy Story de las recomendaciones
peliculas_recomendadas = [pelicula for pelicula in peliculas_recomendadas if pelicula != id_toy_story]

# Mostrar lista filtrada
print(f"Películas recomendadas después de filtrar Toy Story: {peliculas_recomendadas}")

# Obtener detalles de las películas recomendadas
detalles_peliculas_recomendadas = movie[movie['movieId'].isin(peliculas_recomendadas)]
detalles_peliculas_recomendadas.head()
```
